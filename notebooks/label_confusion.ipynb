```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
preds = pd.read_csv('../backend/data/ss-role/fine_tuning_batch_preds.csv')
preds = preds[preds.gold_labels != 'Accs']

gold_labels = preds.gold_labels
epochs = [0, 5, 25, 50, 75, 100, 125, 150, 175, 176]

for epoch in epochs:
    fig, ax = plt.subplots(figsize=(20, 20))
    cm = confusion_matrix(gold_labels, preds[str(epoch)])
    cm_display = ConfusionMatrixDisplay(cm).from_predictions(gold_labels, preds[str(epoch)], ax=ax, xticks_rotation='vertical', normalize='true')
    ax.title.set_text(f'Epoch {epoch}')
    plt.savefig(f'images/confusion_matrices/epoch_{epoch}.png', dpi=300)
```
